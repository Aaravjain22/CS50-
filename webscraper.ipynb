```python
import requests
from bs4 import BeautifulSoup
import json
import time
import random
URLS = [
    "www.google.com",
    "www.youtube.com",
    "www.facebook.com",
    "mail.google.com",
    "docs.google.com",
    "www.yahoo.com",
    "www.amazon.com",
    "twitter.com",
    "en.wikipedia.org",
    "www.reddit.com",
    "www.netflix.com",
    "www.linkedin.com",
    "translate.google.com",
    "www.twitch.tv",
    "search.yahoo.com",
    "www.bing.com",
    "www.microsoft.com",
    "www.ebay.com",
    "mail.yahoo.com",
    "www.google.co.in",
    "www.amazon.co.jp",
    "github.com",
    "www.imdb.com",
    "stackoverflow.com",
    "www.office.com",
    "www.msn.com",
    "www.paypal.com",
    "imgur.com",
    "wordpress.com",
    "www.apple.com",
    "www.adobe.com",
    "www.pinterest.com",
    "www.booking.com",
    "www.tumblr.com",
    "www.dropbox.com",
    "play.google.com",
    "www.roblox.com",
    "www.amazon.co.uk",
    "news.google.com",
    "www.cnn.com",
    "soundcloud.com",
    "www.bbc.co.uk",
    "www.spotify.com",
    "www.salesforce.com",
    "www.chase.com",
    "www.mediafire.com",
    "www.espn.com",
    "www.google.co.uk",
    "www.nytimes.com",
    "www.cricbuzz.com",
    "www.nih.gov",
    "vimeo.com",
    "www.researchgate.net",
    "www.indeed.com",
    "www.etsy.com",
    "www.nlm.nih.gov",
    "www.ncbi.nlm.nih.gov",
    "www.flipkart.com",
    "store.steampowered.com",
    "www.alibaba.com",
    "www.theguardian.com",
    "www.slideshare.net",
    "www.ebay.co.uk",
    "duckduckgo.com",
    "www.walmart.com",
    "www.godaddy.com",
    "news.yahoo.com",
    "www.ask.com",
    "trello.com",
    "www.shutterstock.com",
    "support.microsoft.com",
    "www.hulu.com",
    "www.deviantart.com",
    "www.yelp.com",
    "blogger.com",
    "www.zillow.com",
    "www.sciencedirect.com",
    "www.mozilla.org",
    "indiatimes.com",
    "www.bankofamerica.com",
    "www.wikihow.com",
    "mega.nz",
    "www.scribd.com",
    "speedtest.net",
    "www.espncricinfo.com",
    "www.weather.com",
    "www.cnet.com",
    "archive.org",
    "www.udemy.com",
    "www.wellsfargo.com",
    "www.tripadvisor.com",
    "dailymail.co.uk",
    "www.foxnews.com",
    "www.amazon.ca",
    "rt.com",
    "rt.com",
    "www.ikea.com",
    "www.livejournal.com",
    "www.homedepot.com",
    "www.shopify.com",
    "www.bet365.com",
    "zendesk.com",
    "patreon.com",
    "www.fiverr.com",
    "aws.amazon.com",
    "www.bestbuy.com",
    "9gag.com",
    "www.washingtonpost.com",
    "www.pixabay.com",
    "www.reverso.net",
    "finance.yahoo.com",
    "www.zoho.com",
    "www.gamespot.com",
    "www.leagueoflegends.com",
    "www.behance.net",
    "www.intuit.com",
    "www.usps.com",
    "www.forbes.com",
    "www.americanexpress.com",
    "www.hdfcbank.com",
    "www.hp.com",
    "www.ilovepdf.com",
    "www.capitalone.com",
    "www.samsung.com",
    "www.aol.com",
    "www.siteadvisor.com",
    "www.mailchimp.com",
    "coinmarketcap.com",
    "www.utorrent.com",
    "www.google.com.au",
    "www.elsevier.com",
    "www.wordpress.org",
    "www.accuweather.com",
    "www.asos.com",
    "smallpdf.com",
    "www.irctc.co.in",
    "www.businessinsider.com",
    "www.grammarly.com",
    "www.dell.com",
    "www.cambridge.org",
    "sourceforge.net",
    "www.ign.com",
    "www.ups.com",
    "www.oracle.com",
    "bitly.com",
    "www.springer.com",
    "www.evernote.com",
    "www.flickr.com",
    "www.weebly.com",
    "www.box.com",
    "www.playstation.com",
    "www.goodreads.com",
    "www.myway.com",
    "timesofindia.indiatimes.com",
    "www.fedex.com",
    "www.target.com",
    "www.thesaurus.com",
    "www.investing.com",
    "www.icicibank.com",
    "www.feedly.com",
    "www.discogs.com",
    "www.ndtv.com",
    "www.wordreference.com",
    "www.ultimate-guitar.com",
    "www.nexusmods.com",
    "dictionary.cambridge.org",
    "www.wiley.com",
    "www.mlb.com",
    "www.asus.com",
    "onlinelibrary.wiley.com",
    "www.wowhead.com",
    "answers.yahoo.com",
    "edition.cnn.com",
    "www.skype.com",
    "www.theverge.com",
    "www.go.com",
    "www.usatoday.com",
    "www.nike.com",
    "www.webmd.com",
    "www.billdesk.com",
    "www.rediff.com",
    "www.internetdownloadmanager.com",
    "www.expedia.com",
    "www.adp.com",
    "wiktionary.org",
    "sports.yahoo.com",
    "www.naukri.com",
    "www.hm.com",
    "www.hotels.com",
    "www.webex.com",
    "www.squarespace.com",
    "www.ebay.com.au",
    "www.mit.edu",
    "www.cnbc.com",
    "www.chess.com",
    "www.blackboard.com",
    "www.lowes.com",
    "www.epicgames.com",
    "www.bloomberg.com",
    "archiveofourown.org",
    "www.fidelity.com",
    "www.ca.gov",
    "www.autodesk.com",
    "tweetdeck.twitter.com",
    "developer.mozilla.org",
    "www.att.com",
    "www.surveymonkey.com",
    "www.eventbrite.com",
    "www.reuters.com",
    "www.jw.org",
    "www.japanpost.jp",
    "www.wsj.com",
    "www.techradar.com",
    "www.filehippo.com",
    "www.rakuten.com",
    "www.agoda.com",
    "www.discover.com",
    "www.urbandictionary.com",
    "www.realtor.com",
    "costco.com",
    "www.coursera.org",
    "www.merriam-webster.com",
    "www.nhk.or.jp",
    "www.wunderground.com",
    "download.cnet.com",
    "newegg.com",
    "www.indiamart.com",
    "www.ibm.com",
    "www.khanacademy.org",
    "www.engadget.com",
    "mail.aol.com",
    "www.ieee.org",
    "www.gizmodo.com",
    "www.macys.com",
    "www.intel.com",
    "www.nba.com",
    "www.zhaopin.com",
    "www.kijiji.ca",
    "www.state.gov",
    "www.thefreedictionary.com",
    "www.ea.com"
    "www.bhaskar.com"

]

def clean_text(text):
  return ' '.join(text.strip().split())

def scrape_text(url):
    try:
      headers = {
          "user-agent": "mozilla/5.0"
      }
      response = requests.get(url, headers=headers, timeout=13)
      soup = BeautifulSoup(response.text, 'html.parser') # Corrected 'htam.praser'

      for tag in soup(["script","style","noscript"]):
        tag.decompose()

      text = soup.get_text(separator=' ',strip=True) # Corrected 'seperatpor'
      return clean_text(text)
    except Exception as e: # Corrected 'exception'
      print(f"Error scraping {url}: {e}")
      return ""

def main():
  scraped_data = []
  for idx, url in enumerate(URLS):
    print(f"[{idx+1}/{len(URLS)}]scraping:{url}")
    text = scrape_text(f"http://{url}") # Added http:// and assigned return value to text variable

    if text:
      scraped_data.append({
          "url": url,
          "text": text[:2000]#limits to first 2000 chars
      })

    time.sleep(random.uniform(1.5, 3.5))#sleep to avoid getting blocked

  # JSON saving
  with open("gulli.json", "w", encoding="utf-8") as f:
    json.dump(scraped_data, f, ensure_ascii=False, indent=2)

if __name__ == "__main__":
  main()
```
